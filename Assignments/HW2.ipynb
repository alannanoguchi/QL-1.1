```python
from random import choices

def verify_heads(prob_head, num_flips):
    event = ['H', 'T']
    weights = [prob_head, 1 - prob_head] 
    # the probability of heads is 1/2 so the probability of tails is 1 - 1/2(the probability of heads)

    count = 0
    
    for items in range(num_flips):
        list = [choices(event, weights)[0], choices(event, weights)[0], choices(event, weights)[0]]
        if list in [['H', 'H', 'T']]:
            count += 1
        elif list in [['T', 'H', 'H']]:
            count += 1
        elif list in [['H', 'T', 'H']]:
            count += 1
        else: 
            count += 0
    return count/num_flips # the probability


print("Probability given that P(H) is 0.3 = " + str(verify_heads(0.3, 1000))) # if the coin is not fair
print("Probability given that P(H) is 0.5 = " + str(verify_heads(0.5, 1000))) # if the coin is fair
print("Probability given that P(H) is 0.9 = " + str(verify_heads(0.9, 1000))) # if the coin is not fair
```
